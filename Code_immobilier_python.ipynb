```python
from sklearn.linear_model import LinearRegression
```

```python
df['log_price_m2'] = np.log(df['price_sqm'] + 1)
```

```python
for year in df['anneemut'].unique():
    for month in df['moismut'].unique():
        # Créer un sous-ensemble de données contenant uniquement les observations pour cette année et ce mois
        subset = df[(df['anneemut'] == year) & (df['moismut'] == month) & (df['appart'] == 1) & (df['maison'] == 0)]
        
        # Sélectionner les variables explicatives (toutes les colonnes sauf "log_prix_par_m2", "annee", "mois" et les variables catégorielles)
        X = subset.drop(['log_price_m2','coddep', 'l_codinsee', 'xwgs', 'ywgs', 'price_sqm', 'appart', 
                        'maison','anneemut', 'moismut','DEP_23','garage_0','terrasse_0','T2','cst_1990_2012',
                         'mois_1', 'mois_2', 'mois_3', 'mois_4', 'mois_5', 'mois_6', 'mois_7', 
                         'mois_8', 'mois_9', 'mois_10', 'mois_11', 'mois_12','STATUT_2017','TDUU2017'], axis=1)
        
        # Sélectionner la variable dépendante (la colonne "log_prix_par_m2")
        y = subset['log_price_m2']
        
        # Créer un modèle de régression linéaire
        model = LinearRegression()
        
        # Entraîner le modèle sur les données d'entraînement
        model.fit(X, y)
        
        # Afficher les coefficients de régression et le R²
        print(f'Coefficients de regression pour l\'année {year} et le mois {month}:')
        for i, col_name in enumerate(X.columns):
            print(f'{col_name}: {model.coef_[i]}')
        print(f'R²: {model.score(X, y)}\n')
        
```

```python
def apply_rules(row):
    centre_vars = row.filter(regex='^Centre')
    UU_vars = row.filter(regex='^UU')
    if centre_vars.sum() == 1:
        row.loc[row.filter(regex='^UU').columns] = 0
        row.loc[row.filter(regex='^DEP').columns] = 0
    elif UU_vars.sum() == 1:
        row.loc[row.filter(regex='^DEP').columns] = 0
        row.loc[row.filter(regex='^Centre').columns] = 0
    return row
```

```python
# Application des règles à chaque ligne du dataframe
test = test.apply(apply_rules, axis=1)

# Affichage du résultat
print(test)
```
